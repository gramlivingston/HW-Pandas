```python
import pandas as pd
import numpy as np
```

```python
schools_df = pd.read_csv("Resources/schools_complete.csv")
students_df = pd.read_csv("Resources/students_complete.csv")
schools = pd.merge(students_df, schools_df, how = "left", on=["school_name", "school_name"])
schools['Passed_Reading'] = schools['reading_score'].ge(70)
schools['Passed_Math'] = schools['math_score'].ge(70)
schools.head()
```

```python
schools
```

```python
#District Summary
#Needs to be Cleaned
#Needs to be table
numschools = len(schools['school_name'].value_counts())
size = schools_df['size'].sum()
budget = schools_df['budget'].sum()
avg_math = students_df['math_score'].mean()
avg_reading = students_df['reading_score'].mean()
passed_reading = (len(schools.loc[schools["reading_score"] > 69])/size)*100
passed_math = (len(schools.loc[schools["math_score"] > 69])/size)*100
passed = ((passed_reading+passed_math)/2)
d = {'Total Schools': [numschools], 'Total Students': [size], 'Total Budget': [budget], 'Average Math Score': [avg_math], 'Average Reading Score': [avg_reading], '% Passing Math': [passed_math], '% Passing Reading': [passed_reading], '% Overall Passing Rate':[passed] }
DistSummary = pd.DataFrame(data=d)
DistSummary.head()
```

```python
groupbyschool = schools.groupby(['school_name'])
groupbyschool
```

```python
#Needs to be Cleaned
#GroupBy All students School Chart
groupbyschool = schools.groupby(['school_name'])
passed_reading = ((groupbyschool['Passed_Reading'].sum())/groupbyschool['size'].unique())*100
passed_math = ((groupbyschool['Passed_Math'].sum())/groupbyschool['size'].unique())*100
budgetperstu = (groupbyschool['budget'].unique()/groupbyschool['size'].unique())


byschool = pd.DataFrame({"School Type": groupbyschool['type'].unique(),
                                   "Total Students": groupbyschool['size'].unique(),
                                    "Total Budget": groupbyschool['budget'].unique(),
                                     "Budget per Student": budgetperstu,
                                     "Avg Math Score" : groupbyschool['math_score'].mean(),
                                     "Avg Reading Score" : groupbyschool['reading_score'].mean(),
                                        "% Passed Reading" : groupbyschool['Passed_Reading'].sum()/groupbyschool['size'].unique()*100,
                                        "% Passed Math": passed_math,
                                        "% Passed": (passed_math+passed_reading)/2})
type(byschool)
byschool.head()
```

```python
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
spending_bins = [0, 585, 615, 645, 675]




byschool['byspending'] = pd.cut(byschool['Budget per Student'], bins=spending_bins, labels=group_names)

byspending = byschool.groupby(['byspending'])
byspending.head()

byspending = pd.DataFrame({"Avg Math Score" : byspending['Avg Math Score'].mean(), 
                           "Avg Reading Score" : byspending['Avg Reading Score'].mean() 
                         #  "% Passed Reading" : byspending['% Passed Reading'].mean()
                          })
                           
byspending
```
